```python
import pandas as pd

data = {
    "Регіон": ["Захід", "Захід", "Схід", "Схід", "Південь", "Південь"],
    "Продукт": ["Ноутбук", "Телефон", "Ноутбук", "Телефон", "Ноутбук", "Телефон"],
    "Продажі": [120, 80, 150, 200, 90, 60],
    "Рік": [2024, 2024, 2024, 2024, 2024, 2024]
}

df = pd.DataFrame(data)
print(df)
```

```python
slice_df = df[df["Продукт"] == "Ноутбук"]
print(slice_df)
```

```python
dice_df = df[(df["Регіон"].isin(["Захід","Схід"])) & (df["Продукт"]=="Телефон")]
print(dice_df)
```

```python
rollup_df = df.groupby("Регіон")["Продажі"].sum()
print(rollup_df)
```

```python
pivot = pd.pivot_table(df, values="Продажі", index="Регіон", columns="Продукт", aggfunc="sum")
print(pivot)
```
