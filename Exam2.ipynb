```python
# Базовий генетичний алгоритм
import numpy as np
import random

# Функція пристосованості (максимізація f(x) = x^2)
def fitness(x):
    return x**2

# Параметри
POP_SIZE = 10
GENES = 5   # довжина хромосоми (0..31)
GENERATIONS = 20

# Ініціалізація популяції
population = [np.random.randint(0, 2, GENES) for _ in range(POP_SIZE)]

def decode(chromosome):
    return int("".join(map(str, chromosome)), 2)

def selection(population):
    scores = [fitness(decode(ch)) for ch in population]
    probs = np.array(scores) / sum(scores)
    return population[np.random.choice(range(len(population)), p=probs)]

def crossover(p1, p2):
    point = random.randint(1, GENES-1)
    return np.concatenate([p1[:point], p2[point:]])

def mutation(chromosome, rate=0.1):
    for i in range(len(chromosome)):
        if random.random() < rate:
            chromosome[i] = 1 - chromosome[i]
    return chromosome

# Основний цикл
for g in range(GENERATIONS):
    new_pop = []
    for _ in range(POP_SIZE):
        p1, p2 = selection(population), selection(population)
        child = crossover(p1, p2)
        child = mutation(child)
        new_pop.append(child)
    population = new_pop

best = max(population, key=lambda ch: fitness(decode(ch)))
print("Найкраще рішення:", decode(best), "Значення:", fitness(decode(best)))
```

```python
# Модифікація 1 — турнірний відбір
def tournament_selection(population, k=3):
    selected = random.sample(population, k)
    return max(selected, key=lambda ch: fitness(decode(ch)))
population = [np.random.randint(0, 2, GENES) for _ in range(POP_SIZE)]
best_tourn = max(population, key=lambda ch: fitness(decode(ch)))
print("З турнірним відбором -> Найкраще рішення:", decode(best_tourn),
      "Значення:", fitness(decode(best_tourn)))
```

```python
#Модифікація 2 — багатоточковий кросовер
def crossover(p1, p2):
    point1, point2 = sorted(random.sample(range(1, GENES), 2))
    return np.concatenate([p1[:point1], p2[point1:point2], p1[point2:]])
population = [np.random.randint(0, 2, GENES) for _ in range(POP_SIZE)]
best_two_cross = max(population, key=lambda ch: fitness(decode(ch)))
print("З двоточковим кросовером -> Найкраще рішення:", decode(best_two_cross),
      "Значення:", fitness(decode(best_two_cross)))
```
