```python
def gcd_euclid(a, b):
    if not isinstance(a, int) or not isinstance(b, int):
        raise TypeError("Потрібно вводити лише цілі числа.")
    if a <= 0 or b <= 0:
        raise ValueError("Числа мають бути додатними.")

    while b != 0:
        a, b = b, a % b
    return a
```

```python
print(gcd_euclid(48, 18))   # Очікувано: 6
print(gcd_euclid(101, 10))  # Очікувано: 1
print(gcd_euclid(270, 192)) # Очікувано: 6
```

```python
try:
    print(gcd_euclid("48", 18))
except Exception as e:
    print("Помилка:", e)

try:
    print(gcd_euclid(-5, 15))
except Exception as e:
    print("Помилка:", e)
```

```python
import matplotlib.pyplot as plt

pairs = [(48,18), (101,10), (270,192), (56,98), (120,45)]
results = [gcd_euclid(a,b) for a,b in pairs]

labels = [f"{a},{b}" for a,b in pairs]

plt.figure(figsize=(8,4))
plt.bar(labels, results, color="red")
plt.title("Алгоритм Евкліда: НСД для пар чисел")
plt.ylabel("НСД")
plt.xlabel("Пари чисел")
plt.show()
```
